```python
import mcaa
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams["figure.figsize"] = (16, 5)
```

```python
MUTATION_COUNT = 100
MUTATION_SCALE = np.linspace(0, 1, MUTATION_COUNT)
```

```python
constantSchedule = mcaa.constantSchedule(0.6)
contLinearSchedule = mcaa.continuousLinearSchedule(MUTATION_COUNT, 0.2, 0.7)
contPolynomialSchedule1 = mcaa.continuousPolynomialSchedule(MUTATION_COUNT, 1.1, 0.8)
contPolynomialSchedule2 = mcaa.continuousPolynomialSchedule(MUTATION_COUNT, 1.1, 1.5)
contLogarithmicSchedule = mcaa.continuousLogarithmicSchedule(MUTATION_COUNT, 1.1, 1.2)
contExponentialSchedule = mcaa.continuousExponentialSchedule(MUTATION_COUNT, 0.15, 0.1)
```

```python
constantScheduleValues = [constantSchedule.computeNextTemp() for i in MUTATION_SCALE]
contLinearScheduleValues = [contLinearSchedule.computeNextTemp() for i in MUTATION_SCALE]
contPolynomialScheduleValues1 = [contPolynomialSchedule1.computeNextTemp() for i in MUTATION_SCALE]
contPolynomialScheduleValues2 = [contPolynomialSchedule2.computeNextTemp() for i in MUTATION_SCALE]
contLogarithmicScheduleValues = [contLogarithmicSchedule.computeNextTemp() for i in MUTATION_SCALE]
contExponentialScheduleValues = [contExponentialSchedule.computeNextTemp() for i in MUTATION_SCALE]
```

```python
plt.plot(MUTATION_SCALE, constantScheduleValues, label='constant')
plt.plot(MUTATION_SCALE, contLinearScheduleValues, label='linear')
plt.plot(MUTATION_SCALE, contPolynomialScheduleValues1, label='sublinear')
plt.plot(MUTATION_SCALE, contPolynomialScheduleValues2, label='polynomial')
plt.plot(MUTATION_SCALE, contLogarithmicScheduleValues, label='logarithmic')
plt.plot(MUTATION_SCALE, contExponentialScheduleValues, label='exponential')
plt.title('Temperature over iteration percentage per schedule')
plt.xlabel('Iteration percentage of maximum iterations')
plt.ylabel('Temperature value')
plt.legend(loc='best')
plt.show()
```
