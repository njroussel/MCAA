```python
import mcaa
import matplotlib.pyplot as plt
import numpy as np
```

```python
MUTATION_COUNT = 100
MUTATION_SCALE = np.linspace(0, 1, MUTATION_COUNT)
```

```python
constantSchedule = mcaa.constantSchedule(2.2)
contLinearSchedule = mcaa.continuousLinearSchedule(MUTATION_COUNT, 0.2, 6)
contPolynomialSchedule1 = mcaa.continuousPolynomialSchedule(MUTATION_COUNT, 10, 2.5)
contPolynomialSchedule2 = mcaa.continuousPolynomialSchedule(MUTATION_COUNT, 10, 0.5)
contLogarithmicSchedule = mcaa.continuousLogarithmicSchedule(MUTATION_COUNT, 10, 1)
contExponentialSchedule = mcaa.continuousExponentialSchedule(MUTATION_COUNT, 0.01, 0.0005)
```

```python
constantScheduleValues = [constantSchedule.computeNextTemp() for i in MUTATION_SCALE]
contLinearScheduleValues = [contLinearSchedule.computeNextTemp() for i in MUTATION_SCALE]
contPolynomialScheduleValues1 = [contPolynomialSchedule1.computeNextTemp() for i in MUTATION_SCALE]
contPolynomialScheduleValues2 = [contPolynomialSchedule2.computeNextTemp() for i in MUTATION_SCALE]
contLogarithmicScheduleValues = [contLogarithmicSchedule.computeNextTemp() for i in MUTATION_SCALE]
contExponentialScheduleValues = [contExponentialSchedule.computeNextTemp() for i in MUTATION_SCALE]
```

```python
plt.plot(constantScheduleValues)
plt.plot(contLinearScheduleValues)
plt.plot(contPolynomialScheduleValues1)
plt.plot(contPolynomialScheduleValues2)
plt.plot(contLogarithmicScheduleValues)
plt.plot(contExponentialScheduleValues)
plt.show()
```
