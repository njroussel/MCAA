```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import constantRunner
import linearRunner
import polynomialRunner
import logarithmicRunner
import exponentialRunner
```

```python
plt.rcParams["figure.figsize"] = (16, 5)
```

```python
# Constants

# CONSTANT SCHEDULE
CONST_BETA = 0.6

# LINEAR SCHEDULE
LIN_OFFSET = 0.2
LIN_SLOPE = 0.7

# SUBLINEAR SCHEDULE
SUB_OFFSET = 1.1
SUB_EXPO = 0.8

# POLYNOMIAL SCHEDULE
POLY_OFFSET = 1.1
POLY_EXPO = 1.5

# LOGARITHMIC SCHEDULE
LOG_OFFSET = 1.1
LOG_LOG_OFFSET = 1.2

# EXPONENTIAL SCHEDULE
EXP_OFFSET = 0.15
EXP_BASE = 0.1

ALPHAS = np.arange(0.5, 5.5, 0.5)
N = 100
T_FACTOR = 100 
T_FACTOR = 100 
T = T_FACTOR * N
MEASURE_STEP = 0
RUN_COUNT = 128
THREAD_COUNT = 4
```

```python
CONST_MEASURES = [constantRunner.constantRunner(CONST_BETA, alpha, N, T_FACTOR, MEASURE_STEP, RUN_COUNT, THREAD_COUNT) for alpha in ALPHAS]
LIN_MEASURES = [linearRunner.linearRunner(LIN_OFFSET, LIN_SLOPE, alpha, N, T_FACTOR, MEASURE_STEP, RUN_COUNT, THREAD_COUNT) for alpha in ALPHAS]
SUB_MEASURES = [polynomialRunner.polynomialRunner(SUB_OFFSET, SUB_EXPO, alpha, N, T_FACTOR, MEASURE_STEP, RUN_COUNT, THREAD_COUNT) for alpha in ALPHAS]
POLY_MEASURES = [polynomialRunner.polynomialRunner(POLY_OFFSET, POLY_EXPO, alpha, N, T_FACTOR, MEASURE_STEP, RUN_COUNT, THREAD_COUNT) for alpha in ALPHAS]
LOG_MEASURES = [logarithmicRunner.logarithmicRunner(LOG_OFFSET, LOG_LOG_OFFSET, alpha, N, T_FACTOR, MEASURE_STEP, RUN_COUNT, THREAD_COUNT) for alpha in ALPHAS]
EXP_MEASURES = [exponentialRunner.exponentialRunner(EXP_OFFSET, EXP_BASE, alpha, N, T_FACTOR, MEASURE_STEP, RUN_COUNT, THREAD_COUNT) for alpha in ALPHAS]
```

```python
CONST_E = [measure[0] for measure in CONST_MEASURES]
CONST_OVERLAP = [measure[1] for measure in CONST_MEASURES]
LIN_E = [measure[0] for measure in LIN_MEASURES]
LIN_OVERLAP = [measure[1] for measure in LIN_MEASURES]
SUB_E = [measure[0] for measure in SUB_MEASURES]
SUB_OVERLAP = [measure[1] for measure in SUB_MEASURES]
POLY_E = [measure[0] for measure in POLY_MEASURES]
POLY_OVERLAP = [measure[1] for measure in POLY_MEASURES]
LOG_E = [measure[0] for measure in LOG_MEASURES]
LOG_OVERLAP = [measure[1] for measure in LOG_MEASURES]
EXP_E = [measure[0] for measure in EXP_MEASURES]
EXP_OVERLAP = [measure[1] for measure in EXP_MEASURES]
```

```python
CONST_E_MEAN = np.mean(CONST_E, axis=2)
CONST_E_STD = np.std(CONST_E, axis=2)
LIN_E_MEAN = np.mean(LIN_E, axis=2)
LIN_E_STD = np.std(LIN_E, axis=2)
SUB_E_MEAN = np.mean(SUB_E, axis=2)
SUB_E_STD = np.std(SUB_E, axis=2)
POLY_E_MEAN = np.mean(POLY_E, axis=2)
POLY_E_STD = np.std(POLY_E, axis=2)
LOG_E_MEAN = np.mean(LOG_E, axis=2)
LOG_E_STD = np.std(LOG_E, axis=2)
EXP_E_MEAN = np.mean(EXP_E, axis=2)
EXP_E_STD = np.std(EXP_E, axis=2)
```

```python
CONST_OVERLAP_MEAN = np.mean(CONST_OVERLAP, axis=2)
CONST_OVERLAP_STD = np.std(CONST_OVERLAP, axis=2)
LIN_OVERLAP_MEAN = np.mean(LIN_OVERLAP, axis=2)
LIN_OVERLAP_STD = np.std(LIN_OVERLAP, axis=2)
SUB_OVERLAP_MEAN = np.mean(SUB_OVERLAP, axis=2)
SUB_OVERLAP_STD = np.std(SUB_OVERLAP, axis=2)
POLY_OVERLAP_MEAN = np.mean(POLY_OVERLAP, axis=2)
POLY_OVERLAP_STD = np.std(POLY_OVERLAP, axis=2)
LOG_OVERLAP_MEAN = np.mean(LOG_OVERLAP, axis=2)
LOG_OVERLAP_STD = np.std(LOG_OVERLAP, axis=2)
EXP_OVERLAP_MEAN = np.mean(EXP_OVERLAP, axis=2)
EXP_OVERLAP_STD = np.std(EXP_OVERLAP, axis=2)
```

```python
# Plot
plt.errorbar(ALPHAS, CONST_E_MEAN, yerr=CONST_E_STD, label='constant', capsize=10)
plt.errorbar(ALPHAS, LIN_E_MEAN, yerr=LIN_E_STD, label='linear', capsize=10)
plt.errorbar(ALPHAS, SUB_E_MEAN, yerr=SUB_E_STD, label='sublinear', capsize=10)
plt.errorbar(ALPHAS, POLY_E_MEAN, yerr=POLY_E_STD, label='polynomial', capsize=10)
plt.errorbar(ALPHAS, LOG_E_MEAN, yerr=LOG_E_STD, label='logarithmic', capsize=10)
plt.errorbar(ALPHAS, EXP_E_MEAN, yerr=EXP_E_STD, label='exponential', capsize=10)
plt.title('Average normalized energy at ' + str(T_FACTOR * N) + ' iterations over alpha per schedule')
plt.legend(loc='best')
plt.xlabel('Alpha')
plt.ylabel('Average normalized energy')
plt.show()


        
plt.errorbar(ALPHAS, CONST_OVERLAP_MEAN, yerr=CONST_OVERLAP_STD, label='constant', capsize=10)
plt.errorbar(ALPHAS, LIN_OVERLAP_MEAN, yerr=LIN_OVERLAP_STD, label='linear', capsize=10)
plt.errorbar(ALPHAS, SUB_OVERLAP_MEAN, yerr=SUB_OVERLAP_STD, label='sublinear', capsize=10)
plt.errorbar(ALPHAS, POLY_OVERLAP_MEAN, yerr=POLY_OVERLAP_STD, label='polynomial', capsize=10)
plt.errorbar(ALPHAS, LOG_OVERLAP_MEAN, yerr=LOG_OVERLAP_STD, label='logarithmic', capsize=10)
plt.errorbar(ALPHAS, EXP_OVERLAP_MEAN, yerr=EXP_OVERLAP_STD, label='exponential', capsize=10)
plt.ylim(ymin=0, ymax=1)
plt.legend(loc='best')
plt.title('Average overlap at ' + str(T_FACTOR * N) + ' iterations over alpha per schedule')
plt.xlabel('Alpha')
plt.ylabel('Average overlap')

plt.show()
```
